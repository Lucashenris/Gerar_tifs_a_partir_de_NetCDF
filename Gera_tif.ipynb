```python
import xarray as xr
import pandas as pd
import os
```

```python
#abrir os arquivos .nc
d1 = xr.open_dataset('t2m-diario-2022.nc')
#visualizar conteúdo do arquivo .nc
d1
```

```python
#Range contendo as datas do arquivo .nc a serem convertidas. Obs.: ajuste conforme a sua necessidade
datas = pd.date_range('2021-12-31', '2022-06-01', freq='D')
#transformar a data em string no formato ano-mes-dia
datastr = datas.strftime('%Y-%m-%d')
#mostrar a quantidade de datas
len(datastr)
```

```python
#Laço for para percorrer todas as datas do arquivo .nc
for i in range(1,d1.dims['time']+1):
    #se não existir a pasta, cria
    if not os.path.exists('Tifs'):
        os.makedirs('Tifs')
    #saída do arquivo .tif
    nome = 'Tifs/dados_diarios_t2m_'+datastr[i]+'.tif'
    #usar o comando gdal translate para converter o arquivo .nc em .tif
    os.system(f'gdal_translate -b {i} -a_nodata -999 -of GTiff t2m-diario-2022.nc '+nome)
```
